```python
import pandas as pd
import numpy as np

import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn.metrics import accuracy_score
train=pd.read_csv('/home/deepak/Downloads/hacker earth ml/ml 1/898ce544-0-MLChallenge_1Datac63af4e/train_indessa.csv')
test=pd.read_csv('/home/deepak/Downloads/hacker earth ml/ml 1/898ce544-0-MLChallenge_1Datac63af4e/test_indessa.csv')

```

```python
train=train.fillna(value=0)
train.isnull().any()
test=test.fillna(value=0)
test.isnull().any()
```

```python
#preprocessing 

train.term=train.term.str[:2]
k=pd.get_dummies(train.grade)
frames=[train,k]
train=pd.concat(frames,axis=1)


test.term=test.term.str[:2]
kt=pd.get_dummies(test.grade)
frames=[test,kt]
test=pd.concat(frames,axis=1)
```

```python
train.emp_length.unique()
```

```python
train.emp_length.unique()
qual_dict={"9 years":9,"8 years":8,"10+ years":10,"7 years":7,"6 years":6,"5 years":5,"4 years":4,"3 years":3,"2 years":2,"1 year":1,"< 1 year":0 ,"n/a":0}
train.emp_length=train.emp_length.map(qual_dict)


test.emp_length.unique()
qual_dict={"9 years":9,"8 years":8,"10+ years":10,"7 years":7,"6 years":6,"5 years":5,"4 years":4,"3 years":3,"2 years":2,"1 year":1,"< 1 year":0 ,"n/a":0}
test.emp_length=test.emp_length.map(qual_dict)


```

```python
train['home_ownership'].astype('category')
qual_dictonary={"OWN":6,"MORTGAGE":5,"ANY":4,"NONE":1 , "OTHER":2 ,"RENT": 3}
train.home_ownership=train.home_ownership.map(qual_dictonary)
train.home_ownership.head(20)
train.home_ownership

test['home_ownership'].astype('category')
qual_dictonary={"OWN":6,"MORTGAGE":5,"ANY":4,"NONE":1 , "OTHER":2 ,"RENT": 3}
test.home_ownership=test.home_ownership.map(qual_dictonary)
test.home_ownership.head(20)
```
